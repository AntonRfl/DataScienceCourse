```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

#### Aufgabe 1

```python
def ownKNN(merkmal_x, merkmal_y, target, datenpunkt, k_neighbours):
    neighbours = []
    
    for i in range(len(merkmal_x)):
        dist = np.linalg.norm(merkmal_x[i] - datenpunkt)
        neighbours.append((dist, merkmal_y[i]))
    
    neighbours.sort()
    
    top_k_neighbours = neighbours[:k_neighbours]
    k_values = [n[1] for n in top_k_neighbours]

    return sum(k_values) / k_neighbours
```

#### Aufgabe 2

```python
from sklearn.datasets import make_moons
from sklearn.neighbors import KNeighborsClassifier

X, y = make_moons(n_samples=240, noise=0.1, random_state=42)

merkmal_x = X[:,0]
merkmal_y = X[:,1]
datenpunkt =  (0.4,0.3) 


neighbour_1 = ownKNN(merkmal_x, merkmal_y, datenpunkt, k_neighbours=5)
neighbour_1
```

```python
neighbour_2 = KNeighborsClassifier(n_neighbors=5)
neighbour_2.fit(X, y)
print(neighbour_2.predict([datenpunkt]))
```
