# Lineare Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generierung eines künstlichen Datensatzes

np.random.seed(42)  #macht das Beispiel reproduzierbar
m = 100  # Anzahl Instanzen
X = 5 * np.random.rand(m, 1)  # Spaltenvektor
y = 6 - 2 * X + np.random.randn(m, 1)  # Spaltenvektor
print(X, X.shape)
print(y, y.shape)
```

```python
# Plotten des Datensatzes

plt.plot(X, y, "b.")
```

```python
# Offset in den Datensatz einfügen

X_offset = np.ones((X.shape[0], 2))
print(X_offset)

X_offset[:, 1:2] = X
X_offset
```

```python
# Berechnung der linearen Regression mittels Normalengleichung

theta_best = np.linalg.inv(X_offset.T @ X_offset) @ X_offset.T @ y
theta_best
```

```python
# Plotten des Datensatzes mit Approximation

plt.plot(X, y, "b.")
plt.plot(X_offset, theta_best[1]*X_offset + theta_best[0], "r-")
plt.show()
```

```python
# Den Datensatz mit Offset direkt über sklearn
from sklearn.preprocessing import add_dummy_feature

X_offset = add_dummy_feature(X)  # füge x0 = 1 zu jedem Datenpunkt hinzu
X_offset
```

# Lineare Regression mit Scikit-Learn

```python

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
lin_reg.intercept_ # intercept ist Offsetterm
lin_reg.coef_ # coeff sind die Koeffizienten
print(lin_reg.coef_, lin_reg.intercept_)
```

```python
# Vorhersage mit neuen Daten

lin_reg.predict(np.array([[1]])) # Koordinate x = 1 als Input
```

```python
# R2-Score berechnen

lin_reg.score(X, y)
```

```python
y_pred = lin_reg.predict(X)
y_pred.shape
```

```python
plt.scatter(X, y, marker='.', alpha=0.5, color='blue')
plt.scatter(X, y_pred, marker='.', alpha=0.5, color='red')
plt.xlim(0, 4)
plt.ylim(0,10)
```

# Batch-Gradientenverfahren

```python

eta = 0.1  # Lernrate
n_epochs = 1000

m = len(X_offset)  # Anzahl Instanzen

np.random.seed(42)
theta_batch = np.random.randn(2, 1)  # zufällig initialisierte Modellparameter

for epoch in range(n_epochs):

    gradients = 2 / m * X_offset.T @ (X_offset @ theta_batch - y)
    theta_batch = theta_batch - eta * gradients

print(theta_batch)
```

```python
# Plotten des Datensatzes mit Approximation

plt.plot(X, y, "b.")
plt.plot(X, theta_batch[1]*X + theta_batch[0], "r-")
plt.show()
```

# Stochastic Gradient Descent

```python
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor()
sgd_reg.fit(X, y)
sgd_reg.intercept_, sgd_reg.coef_
```

```python
# Plotten des Datensatzes mit Approximation

theta_stoch = np.append(sgd_reg.intercept_, sgd_reg.coef_)
plt.plot(X, y, "b.")
plt.plot(X, theta_stoch[1]*X + theta_stoch[0], "r-")
plt.show()
```

```python
sgd_reg.score(X, y)
```

```python
sgd_reg = SGDRegressor(max_iter = 1000, tol = 1e-3, penalty = None, eta0 = 0.1)
sgd_reg.fit(X, y)

sgd_reg.score(X, y)
```

# Polynomielle Regression

```python
# # Generierung eines künstlichen Datensatzes


np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X ** 2 + X + 2 + np.random.randn(m, 1)

```

```python
# Plotten des Datensatzes

plt.plot(X, y, "b.")
```

```python
# Hinzufügen des quadratischen Merkmals


from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
print(X_poly)
```

```python
# Lineare Regression auf erweiterten Trainingsdaten

poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
poly_reg.intercept_, poly_reg.coef_
```

```python
# Plotten des Datensatzes mit Approximation

theta_poly = np.append(poly_reg.intercept_, poly_reg.coef_)
y_poly = theta_poly[2]*(X**2) + theta_poly[1]*X + theta_poly[0]

plt.plot(X, y, "b.")
plt.plot(X, y_poly, "r.")
plt.show()
```

# Lernkurven

```python
from sklearn.model_selection import learning_curve

train_sizes, train_scores, valid_scores = learning_curve(LinearRegression(), X, y, train_sizes=np.linspace(0.01, 1.0, 40), cv=5, scoring="neg_root_mean_squared_error") #cv = Croos validation

train_errors = -train_scores.mean(axis=1)
valid_errors = -valid_scores.mean(axis=1)

plt.plot(train_sizes, train_errors, "r-+", linewidth=2, label="Trainingsdatensatz")
plt.plot(train_sizes, valid_errors, "b-", linewidth=3, label="Validierungsdatensatz")
plt.show()
```

```python
from sklearn.pipeline import make_pipeline

polynomial_regression = make_pipeline(PolynomialFeatures(degree=10, include_bias=False),LinearRegression())
train_sizes, train_scores, valid_scores = learning_curve(polynomial_regression, X, y, train_sizes=np.linspace(0.01, 1.0, 40), cv=5,scoring="neg_root_mean_squared_error")

train_errors = -train_scores.mean(axis=1)
valid_errors = -valid_scores.mean(axis=1)

plt.plot(train_sizes, train_errors, "r-+", linewidth=2, label="Trainingsdatensatz")
plt.plot(train_sizes, valid_errors, "b-", linewidth=3, label="Validierungsdatensatz")
plt.ylim(0, 5)
plt.show()
```

# Logistische Regression

```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target
print(y)
print(X.shape, y.shape)
```

```python
import sklearn.linear_model as lm 

logr = lm.LogisticRegression(max_iter = 1000)
logr.fit(X,y)
logr.score(X,y)
```

```python
n_classes = 3
colors = 'bwr'
CMAP = colors # Farbpalette
plot_step = 0.01

fig = plt.figure(1, figsize=(18, 9))

# Berücksichtigung von paarweisen Korrelationen 
pair = [2,3]
X = iris.data[:, pair]
y = iris.target

clf = logr.fit(X, y)

# Achsen und Größe des Subplots
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                        np.arange(y_min, y_max, plot_step))

# Labeln des Subplots
plt.xlabel(iris.feature_names[pair[0]])
plt.ylabel(iris.feature_names[pair[1]]) 
plt.axis("tight")

# Pltten der ausgewerteten Datenpunkte
for i, color in zip(range(n_classes), colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=color, edgecolor="black", lw=2, label=iris.target_names[i])
        

    plt.axis("tight")
    plt.legend(loc="upper left")
```
