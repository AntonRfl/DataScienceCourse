```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pylab as plt

RNDST = np.random.seed(52)
CMAP = 'bwr'
fig = plt.figure(1, figsize=(9, 6))

X, y = make_blobs(n_samples=200, centers=2, random_state=RNDST, cluster_std=5.)
plt.scatter(X[:, 0], X[:, 1], s=80, c=y, cmap=CMAP)

plt.xlabel('x')
plt.ylabel('y')
```

```python
from sklearn.tree import DecisionTreeClassifier
own_tree = DecisionTreeClassifier(max_depth=4)
own_tree.fit(X, y)
print(own_tree.score(X,y))
```

```python
fig = plt.figure(1, figsize=(9, 6))
ax = plt.gca()
    
ax.scatter(X[:, 0], X[:, 1], c=y, s=80, cmap=CMAP, clim=(y.min(), y.max()), zorder=3)
ax.axis('tight')
xlim = ax.get_xlim()
ylim = ax.get_ylim()
    
xx, yy = np.meshgrid(np.linspace(*xlim, num=200),np.linspace(*ylim, num=200))
Z = own_tree.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
n_classes = len(np.unique(y))
contours = ax.contourf(xx, yy, Z, alpha=0.7,
                       levels=np.arange(n_classes + 1) - 0.5,
                       cmap=CMAP, clim=(y.min(), y.max()),
                       zorder=1)

ax.set(xlim=xlim, ylim=ylim)
plt.xlabel('x')
plt.ylabel('y')
```

```python
from sklearn import tree

tree.plot_tree(own_tree)
```

Die Klasse DecisionTreeClassifier bietet einige weitere Parameter, die die Form des Entscheidungsbaums in ähnlicher Weise einschränken:

- max_depth: Maximale Tiefe des Entscheidungsbaums

- min_samples_leaf: Minimale Anzahl an Datenpunkten, die ein Blatt haben muss.

- min_samples_split: Minimale Anzahl an Datenpunkten, die ein Knoten aufweisen muss, damit er aufgeteilt werden kann.

- max_leaf_nodes: Maximale Anzahl Blätter.

- criterion: Kriterium, welches zur Berechnung des Entscheidungsbaums herangezogen wird.


Hier nicht verwendete Merkmale:

- max_features: Maximale Anzahl der Merkmale, die beim Aufteilen eines Knoten berücksichtigt werden.

- min_weight_fraction_leaf: Wie min_samples_leaf, aber als Anteil der gesamten gewichteten Datenpunkte.

```python
own_tree = DecisionTreeClassifier(max_depth=5, min_samples_leaf=3, min_samples_split=2, max_leaf_nodes=8)
#own_tree = DecisionTreeClassifier(max_depth=5,min_samples_leaf=1,min_samples_split=2, max_leaf_nodes=None, criterion='entropy')
own_tree.fit(X, y)
print(own_tree.score(X,y))
```

```python
fig = plt.figure(1, figsize=(9, 6))
ax = plt.gca()
    
ax.scatter(X[:, 0], X[:, 1], c=y, s=80, cmap=CMAP, clim=(y.min(), y.max()), zorder=3)
ax.axis('tight')
xlim = ax.get_xlim()
ylim = ax.get_ylim()
    
xx, yy = np.meshgrid(np.linspace(*xlim, num=200),np.linspace(*ylim, num=200))
Z = own_tree.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
n_classes = len(np.unique(y))
contours = ax.contourf(xx, yy, Z, alpha=0.7,
                       levels=np.arange(n_classes + 1) - 0.5,
                       cmap=CMAP, clim=(y.min(), y.max()),
                       zorder=1)

ax.set(xlim=xlim, ylim=ylim)
plt.xlabel('x')
plt.ylabel('y')
```

```python
tree.plot_tree(own_tree)
```

# Iris-Datensatz

```python
from sklearn.datasets import load_iris
iris = load_iris()

X = iris.data
y = iris.target

own_tree = DecisionTreeClassifier(max_depth=2)
#own_tree = DecisionTreeClassifier()
own_tree.fit(X, y)
print(own_tree.score(X,y))
```

```python
n_classes = 3
colors = 'bwr'#['b','y','r']
plot_step = 0.01

fig = plt.figure(1, figsize=(18, 9))

for pairidx, pair in enumerate([[0, 1], [0, 2], [0, 3],
                                [1, 2], [1, 3], [2, 3]]):
    X = iris.data[:, pair]
    y = iris.target

    clf = own_tree.fit(X, y)

    plt.subplot(2, 3, pairidx + 1)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=CMAP, alpha=0.5)

    plt.xlabel(iris.feature_names[pair[0]])
    plt.ylabel(iris.feature_names[pair[1]])
    plt.axis("tight")

    for i, color in zip(range(n_classes), colors):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c=color, edgecolor='black', lw=2, label=iris.target_names[i])

    plt.axis("tight")

plt.legend(loc='upper left')
```
