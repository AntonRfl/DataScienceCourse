#### Aufgabe 1

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
import pandas as pd
import numpy as np

model_gauss = GaussianNB()

wine = load_wine(as_frame=True)
wine_df = pd.DataFrame(wine.data)
X = wine.data
y = wine.target


# a) Teilen die den Datensatz in eine zwei gleich große Test- und Trainingsmengen.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# b) Trainieren Sie einen Gauß-Naive-Bayes-Klassifizierer mit der Methode GaussianNB aus der Klasse sklearn.naive_bayes.
model_gauss.fit(X_train, y_train)
model_gauss.score(X_test, y_test)
y_pred = model_gauss.predict(X_test)
confusion_matrix_wine = confusion_matrix(y_test, y_pred)

# Extra: normalisierte Matrix ohne Plot
confusion_matrix_normalized = confusion_matrix(y_test, y_pred, normalize='true')
print(confusion_matrix_wine, "\n", confusion_matrix_normalized)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

# c) Bestimmen Sie Konfusionsmatrix f ̈ur die Trainings- und f ̈ur die Testmenge

# Plot non-normalized confusion matrix
titles_options = [
    ("Confusion matrix, without normalization", None),
    ("Normalized confusion matrix", "true"),
]
for title, normalize in titles_options:
    disp = ConfusionMatrixDisplay.from_estimator(
        model_gauss,
        X_test,
        y_test,
        display_labels=list(load_wine().target_names),
        cmap=plt.cm.Blues,
        normalize=normalize,
    )
    disp.ax_.set_title(title)

    print(title)
    print(disp.confusion_matrix)

plt.show()
```

#### Aufgabe 2
Importieren Sie aus scikit-learn den Datensatz sklearn.datasets.make_moons. Generieren Sie einen
Menge aus 1000 Datenpunkten und setzen Sie den Parameter noise=0.08.

```python
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix

# a) Wiederholen Sie die Teilaufgaben a)-c) aus der vorherigen Aufgabe für diesen Datensatz.
X, y = make_moons(1000, noise=0.08, random_state=42)

# a.1) Teilen die den Datensatz in eine zwei gleich große Test- und Trainingsmengen.
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.5, random_state = 42, shuffle=True)

# a.2) Trainieren Sie einen Gauß-Naive-Bayes-Klassifizierer mit der Methode GaussianNB aus der Klasse sklearn.naive_bayes.
model_moon = GaussianNB()
model_moon.fit(X_train, y_train)
model_moon.score(X_train, y_train)
y_pred = model_moon.predict(X_test)

# a.3) Bestimmen Sie Konfusionsmatrix für die Trainings- und für die Testmenge
confusion_matrix_moon = confusion_matrix(y_test, y_pred)
print(confusion_matrix_moon)
```

```python
# Plot non-normalized confusion matrix
titles_options = [
    ("Confusion matrix, without normalization", None),
    ("Normalized confusion matrix", "true"),
]
for title, normalize in titles_options:
    disp = ConfusionMatrixDisplay.from_estimator(
        model_moon,
        X_test,
        y_test,
        display_labels=[0,1],
        cmap=plt.cm.Blues,
        normalize=normalize,
    )
    disp.ax_.set_title(title)

    print(title)
    print(disp.confusion_matrix)

plt.show()
```
